# Dati Strutturati

I **dati strutturati** sono informazioni organizzate in un formato predefinito e rigido, tipicamente una tabella composta da righe (osservazioni) e colonne (variabili o feature).
Ogni colonna ha un tipo di dato ben definito (es. intero, testo, data). In Python, la libreria principale per manipolare dati strutturati è **Pandas**, che mette a disposizione l'oggetto `DataFrame`.


```python
import pandas as pd
```

```python
BASE_URL = "https://github.com/ProfAI/machine-learning-fondamenti/blob/main/datasets/"
```

### CSV

Il formato **CSV (Comma-Separated Values)** è uno dei formati più semplici e diffusi per lo scambio di dati tabulari. Ogni riga del file rappresenta una riga della tabella, e i valori delle colonne sono separati da una virgola `,`.
- **Vantaggi**: Leggero, leggibile come testo semplice, supportato universalmente.
- **Svantaggi**: Non supporta tipi di dati complessi o formule, non ha uno standard rigoroso per la gestione di caratteri speciali (es. virgole nel testo).


```python
df = pd.read_csv(BASE_URL+"shirts_example.csv", index_col=0)
df.head()
```

### TSV

Il formato **TSV (Tab-Separated Values)** è del tutto analogo al CSV, ma utilizza una tabulazione (`\t`) come separatore anziché una virgola.
- **Perché usarlo**: È particolarmente utile quando i dati di testo all'interno delle colonne contengono già virgole, evitando errori di parsing senza ricorrere a complessi sistemi di virgolette.


```python
df = pd.read_csv(BASE_URL+"shirts_example.tsv", index_col=0, sep="\t")
df.head()
```

### EXCEL
Formati supportati xls, xlsx, xlsm, xlsb, odf, ods e odt.

I file **Excel** (come `.xlsx` o il formato open source `.ods` usato in questo esempio) sono formati binari più complessi rispetto ai file di testo come CSV/TSV. Possono contenere più fogli di lavoro, formattazioni, formule e grafici.
- **In Pandas**: Per leggere questi file usiamo `pd.read_excel`. Spesso è necessario installare motori di terze parti (come `odfpy` per `.ods` o `openpyxl` per `.xlsx`) per consentire a Pandas di decodificare il file binario.


```python
pip install odfpy
```

```python
df = pd.read_excel(BASE_URL+"shirts_example.ods", index_col=0)
df.head()
```

#### HTML

Le pagine **HTML** contengono tabelle definite dal tag `<table>`. Pandas permette di effettuare uno scraping automatico di queste tabelle tramite la funzione `pd.read_html`, che analizza il codice HTML e restituisce una lista di DataFrame corrispondenti a tutte le tabelle trovate nella pagina.


```python
df = pd.read_html(BASE_URL+"shirts_example.html", index_col=0, header=0)
df[0].head()
```

### XML

**XML (Extensible Markup Language)** è un linguaggio di marcatura gerarchico che utilizza tag personalizzati per definire la struttura dei dati. È molto comune per lo scambio di informazioni sul web.
- **In Pandas**: A partire dalla versione 1.3.0, Pandas ha introdotto la comoda funzione `pd.read_xml` per convertire direttamente strutture XML nidificate in tabelle bidimensionali.


```python
# serve la version 1.3.0
!pip install pandas --upgrade # --user # solo per l'utente corrente
```

```python
df = pd.read_xml(BASE_URL+"shirts_example.xml")
df.head()
```

### JSON

**JSON (JavaScript Object Notation)** è il formato standard per le API web e lo scambio di dati asincrono. È strutturato come coppie chiave-valore (molto simile ai dizionari Python) e supporta strutture annidate.
- **In Pandas**: La funzione `pd.read_json` permette di caricare file JSON. Se la struttura è piatta, la conversione in DataFrame è immediata; per strutture complesse e nidificate si può usare `pd.json_normalize`.


```python
df = pd.read_json(BASE_URL+"shirts_example.json")
df.head()
```
